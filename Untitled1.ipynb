```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score
loan_pred_df = pd.read_csv('loan_approval_ds.csv')
loan_pred_df.head()
```

```python
loan_pred_df.dtypes
```

```python
loan_pred_rej = loan_pred_df[loan_pred_df['loan_status'] == 'Rejected']
loan_pred_app = loan_pred_df[loan_pred_df['loan_status'] == 'Approved']
```

```python
loan_pred_app['self_employed'].value_counts()
```

```python
loan_pred_app['bank_asset_value'].mean()
```

```python
X = loan_pred_df[['no_of_dependents', 'education', 'self_employed', 'income_annum', 'loan_amount', 'loan_term', 'cibil_score', 'residential_assets_value', 'commercial_assets_value', 'luxury_assets_value', 'bank_asset_value']]
y = loan_pred_df[['loan_status']]
```

```python
categorical_features = ['education', 'self_employed']
numerical_features = ['no_of_dependents', 'income_annum', 'loan_amount', 'loan_term', 'cibil_score', 'residential_assets_value', 'commercial_assets_value', 'luxury_assets_value', 'bank_asset_value']
```

```python
# One-hot encode the categorical variables
encoder = OneHotEncoder(drop='first', sparse_output=False)
X_encoded = encoder.fit_transform(X[categorical_features])

# Standardize numerical features (optional but often useful for models like RandomForest)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X[numerical_features])

# Combine processed categorical and numerical features
X_processed = np.hstack([X_scaled, X_encoded])
```

```python
with open('encoder.pkl', 'wb') as f:
    pickle.dump(encoder, f)
    
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
# Splitting the data into train and test sets
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# Initialize the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
cv_results = cross_val_score(rf_model, X_processed, y, cv=kf, scoring='accuracy')

# Output the accuracy for each fold and the mean accuracy
print(f'Cross-validation accuracies for each fold: {cv_results}')
print(f'Mean accuracy: {cv_results.mean()}')
```

```python
rf_model.fit(X_processed, y)
```

```python
# Predicting on the test set
y_pred = rf_model.predict(X_processed)
```

```python
# Evaluate the model
accuracy = accuracy_score(y, y_pred)
report = classification_report(y, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
with open('loan_pred_model.pkl', 'wb') as f:
    pickle.dump(rf_model, f)
```
